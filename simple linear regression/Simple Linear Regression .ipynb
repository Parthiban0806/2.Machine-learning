```python
#Importing panda  library to read data from database
import pandas as pd
```

```python
#read_csv file is to read the data in our file
dataset=pd.read_csv("salary_Data.csv")
```

```python
dataset
```

```python
#creating a variable for input data
independent=dataset[["YearsExperience"]]
independent
```

```python
#creating a variable for output data

dependent=dataset[["Salary"]]
dependent
```

```python
#matplotlib it contain class and function for draw a graph

import matplotlib.pyplot as plt
```

```python
#

plt.scatter(independent,dependent)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
#sklearn is a class it contains n number of function,in this we split training set and data set 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#sklearn is a class in that we have linear regression function for predict the one input and output.
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#to check weight value from the predicted data

weight=regressor.coef_

weight
```

```python
##to check bais value from the predicted data
bais=regressor.intercept_
bais
```

```python
#to predict the output value based on the input training datas
y_pred=regressor.predict(x_test)
```

```python
#to check the score for the dateset
from sklearn.metrics import r2_score
r=r2_score(y_test,y_pred)
```

```python
r
```

```python
#pickle is a library for save the file
import pickle
filename="finalmodel.sav"
pickle.dump(regressor,open(filename,'wb'))
```
